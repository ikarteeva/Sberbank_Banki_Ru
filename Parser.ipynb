## **Парсер отзывов сотрудников Сбербанка с banki.ru**

```python
from bs4 import BeautifulSoup
import lxml
import requests
```

```python
url = 'https://www.banki.ru/services/official/bank/?ID=322&PAGEN_1='
```

### *Основной блок*

```python
itog = []

i=1

while i < 242:
  url1=url+str(i)
  r = requests.get(url1)
  soup = BeautifulSoup(r.content, from_encoding=r.encoding)
  quotes = soup.find_all('td', class_='article-text', style="word-break: break-word;")
  i += 1

  for element in quotes:
    itog.append(element.get_text())
```

```python
print("Пример отзыва:")
print(itog[0])
```

```python
print(len(itog)) 
```

### *Вывод в файл*

```python
f = open( 'output.txt', 'w' )
for item in itog:
    f.write("%s\n" % item)
f.close()
```

### *Блок подсчета информации*

Обращения/Жалобы, связанные со справками:

```python
schetglobal = 0
for item in itog:
  schet=0
  if(item.__contains__("справк")):
    schet = 1
  schetglobal += schet 
print(schetglobal)
```

Обращения/Жалобы, связанные с декретами:

```python
schetglobal = 0
for item in itog:
  schet=0
  if(item.__contains__("декрет")):
    schet = 1
  schetglobal += schet 
print(schetglobal)
```
